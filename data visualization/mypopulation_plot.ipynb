### 우리 동네 인구 구조 시각화

- 사용한 csv 파일 : incheon_population.csv
- 인천 내 지역, 연수 1동 조사

#### 결과 분석

- 50-60대가 가장 많은 인구 분포를 가지고 있음
- 그 다음으로는 20-30세의 인구가 많음

```python
import csv
import matplotlib.pyplot as plt

f = open('incheon_population.csv', 'r')
data = csv.reader(f)
header = next(data)

result = []
for row in data:
    if '연수1동' in row[0]:
        for m,f in zip(row[3:104], row[106:207]):
            result.append(int(m)+int(f))
print(result)
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title('연수 1동 연령별 인구수')
plt.style.use('ggplot')
plt.plot(result)
plt.xlabel('연령')
plt.ylabel('인구수')
plt.show()
```
