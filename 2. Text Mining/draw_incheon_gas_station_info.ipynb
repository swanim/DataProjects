## 인천광역시의 주유소 데이터 결과 해석

```python
from glob import glob
import pandas as pd
```

```python
station_files = glob('지역*.xls')
station_files
```

```python
import pandas as pd
tmp_list = []

for file_name in station_files:
    tmp = pd.read_excel(file_name, header = 2)
    tmp_list.append(tmp)

"""for file_name in station_files:
    tmp = pd.read_excel(file_name, header = 2)
    tmp_list.append(tmp)"""

tmp_list[0]
```

```python
stations = pd.concat(tmp_list)
stations.shape
```

```python
stations_sel = stations.loc[:,['상표', '상호', '주소', '셀프여부', '휘발유']]
stations_sel
```

```python
stations_sel.columns = ['상표', '주유소', '주소', '셀프여부', '가격']
stations_sel
```

```python
g_list = []

for address in stations['주소']:
    g_list.append("휘발유")
    
stations_sel['종류'] = g_list
stations_sel
```

```python
gu_list = []

for address in stations['주소']:
    address_gu = address.split()[1]
    gu_list.append(address_gu)
    
stations_sel['군/구'] = gu_list
stations_sel
```

```python
stations_sel = stations_sel.query("가격 != '-'").copy()
stations_sel

stations_sel2 = stations_sel2.query("가격 != '-'").copy()
stations_sel2
```

# 그래프 결과 해석

### 1. 그래프 1

- 경유는 평균적으로 1900원 정도의 가격에 분포해 있고, 휘발유는 평균적으로 1600-1700원 사이의 가격에 분포해 있다. 따라서 경유보다 휘발유가 전반적으로 싸다는 것을 확인할 수 있다. 

- 휘발유와 경유 둘 다 셀프일 경우에 가격이 좀 더 싸다.

### 2. 그래프 2

- 어느 상표이든지간에, 휘발유가 경유보다 싸다.

- 알뜰주유소가 가장 싸고 자가상표가 가장 비싸다.

- 휘발유의 가격은 대체적으로 1600원 선에 분포해 있고, 경유의 가격은 대체적으로 1900원 선에 분포해 있다.

### 3. 그래프 3

- 옹진군의 휘발유와 경유 가격이 다른 군/구에 비해 비싸다. 

- 대체적으로 휘발유의 가격이 경유보다 더 싸다.
