### 동계 올림픽 메달 개수 시각화하기

- 동계 올림픽 메달 집계 데이터를 참가 횟수와 금메달 개수로 정렬하여 상위 15개국을 막대 그래프로 시각화하기
    - 참가 횟수, 금메달 개수 순서로 내림차순 정렬

```python
import pandas as pd
```

```python
url = "https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84"
```

```python
df = pd.read_html(url)
df
```

```python
# 슬라이싱을 활용하여 동계올림픽에 대한 데이터 추출

a = df[0].iloc[:,6:11]
a
```

```python
# 컬럼명 설정하기
a.columns = ['동계 참가 횟수','금','은','동','계']
a
```

```python
# 참가횟수, 금메달 기준으로 내림차순 정렬하기
a = a.sort_values(by=['동계 참가 횟수','금'],ascending=False)
a
```

### 데이터 시각화하기

```python
#matplotlib 설정하기
import matplotlib.pyplot as plt

plt.rc('font',family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.style.use("ggplot")
```
